```python
# Import Libraries

import pandas as pd
import seaborn as sns
import matplotlib 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
#used to adjust the configuration of the plots
```

```python
# Read the data
df=pd.read_csv('movies.csv')

# View the data 'movies'
df.head()
```

```python

# Find the missing data 
for col in df.columns:
    missing_val= np.mean(df[col].isnull())
    print(col,'-',round(missing_val*100))
```

```python
# To find the data type of each column
print(df.dtypes)
```

```python
# Changing the data type of budget & gross to 'int'
df["budget"] = df["budget"].fillna(0).astype('int64')
df["gross"] = df["gross"].fillna(0).astype('int64')
print(df.dtypes)
#We need to replace NaN values because NaNs cannot be converted to integers. Converting columns with NaNs directly to int64 will result in an error. 
#Replacing NaNs with a placeholder (like 0) allows the conversion to succeed.
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
#sorting based on the gross value
dff=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicate
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
#scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])# x=budget & y=gross (axis)

plt.title('Budget Vs Gross') #Declaring the Graph Title

plt.xlabel('Gross Earning') # Labeling the x axis as Gross earnings

plt.ylabel('Budget for Film') # Labeling the y axis as Budget
plt.show()
```

```python
dff.head()
```

```python
# plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=dff, scatter_kws={"color": "orange"}, line_kws={"color": "green"})
# We did some formatting and added color to differentiate between scatter and line 
```

```python
# Correlation
# To do correlation, we need to select only numeric columns
# There are 3 types of Correlation. Now we use Pearson. The other 2 are Kendall and Spearson.
num_df=dff.select_dtypes(include=['number'])
correlation_matrix = num_df.corr()

print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation_matrix') #Declaring the Graph Title

plt.xlabel('Move') # Labeling the x axis as Gross earnings

plt.ylabel('Move') # Labeling the y axis as Budget
plt.show()
```

```python
# based on company
dff.head()
```

```python
# To change object-type columns to categorical and then assign numeric codes to these categories.
df_numerized = dff
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
sns.stripplot(x="rating", y="gross", data=dff)
```

```python
num_df=df_numerized.select_dtypes(include=['number'])
correlation_matrix = num_df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation_matrix') #Declaring the Graph Title

plt.xlabel('Move') # Labeling the x axis as Gross earnings

plt.ylabel('Move') # Labeling the y axis as Budget
plt.show()
```

```python
corr_pair = correlation_matrix.unstack()
print(corr_pair)
```
